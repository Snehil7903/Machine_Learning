```python
df.replace("?", np.nan, inplace=True)
```

```python
print(df.isnull().sum())
```

```python
missing = df.isnull().sum()

missing_percent = (missing / len(df)) * 100

print(pd.DataFrame({
    "Missing Values": missing,
    "Percentage": missing_percent
}))
```

```python
df["workclass"] = df["workclass"].fillna(
    df["workclass"].mode()[0]
)
```
